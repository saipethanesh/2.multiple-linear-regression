```python
import pandas as pd #pandas=create and maintance database function, pd is called as pandas 
```

```python
dataset=pd.read_csv("Salary_Data.csv")  # get input from us
```

```python
dataset  # print the database
```

```python
independent=dataset[["YearsExperience"]] # split entier cloumn as input
```

```python
independent # print input
```

```python
dependent=dataset[["Salary"]] # split entier cloumn as output
```

```python
dependent # output print
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split( independent, dependent, test_size=0.30,random_state=0)  #split entier rowand cloumn as input and output and train , model the database
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train) # train model function
```

```python
weight=regressor.coef_
weight # finding slope value 
```

```python
bais=regressor.intercept_
bais # finding the bais value
```

```python
y_pred=regressor.predict(x_test) # traing model
```

```python
from sklearn.metrics import r2_score 
r_score=r2_score(y_test,y_pred)  # checking traning and model
```

```python
r_score  # print score value model value >= 1 is good model
```

```python
import pickle
filename="finalized_model_linear.sav" # store files 
```

```python
pickle.dump(regressor,open(filename, 'wb')) # this function is used open saved files
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb'))
result=loaded_model.predict([[15]]) # loader files and give input and test it 
```
